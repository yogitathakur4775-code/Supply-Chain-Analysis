# STEP 12 — Matplotlib Visualization

```python
# Revenue by Product Type
revenue = df.groupby("Product type")["Revenue generated"].sum()

plt.figure(figsize=(6,4))
plt.bar(revenue.index, revenue.values)
plt.title("Revenue by Product Type")
plt.xlabel("Product Type")
plt.ylabel("Revenue")
plt.show()
```

```python
# Chart 2
# Average Price
price = df.groupby("Product type")["Price"].mean()

plt.figure(figsize=(6,4))
plt.bar(price.index, price.values)
plt.title("Average Price by Product Type")
plt.show()
```

```python
# Chart 3
# Transportation Mode Count
transport = df["Transportation modes"].value_counts()

plt.figure(figsize=(6,4))
plt.pie(transport.values, labels=transport.index, autopct="%1.1f%%")
plt.title("Transportation Modes")
plt.show()
```

```python
# Chart 4
# Revenue Distribution
plt.figure(figsize=(7,4))
plt.hist(df["Revenue generated"], bins=10)
plt.title("Revenue Distribution")
plt.xlabel("Revenue")
plt.ylabel("Frequency")
plt.show()
```

```python
# Chart 5
# Price vs Revenue
plt.figure(figsize=(6,5))
plt.scatter(df["Price"], df["Revenue generated"])
plt.xlabel("Price")
plt.ylabel("Revenue")
plt.title("Price vs Revenue")
plt.show()
```

```python
# Chart 6
# Shipping Cost by Carrier
shipping = df.groupby("Shipping carriers")["Shipping costs"].mean()

plt.figure(figsize=(7,4))
plt.bar(shipping.index, shipping.values)
plt.title("Average Shipping Cost")
plt.show()
```

```python
# Chart 7
# Supplier Manufacturing Cost

supplier = df.groupby("Supplier name")["Manufacturing costs"].mean()

plt.figure(figsize=(8,5))
plt.bar(supplier.index, supplier.values)
plt.title("Manufacturing Cost by Supplier")
plt.xticks(rotation=45)
plt.show()
```
